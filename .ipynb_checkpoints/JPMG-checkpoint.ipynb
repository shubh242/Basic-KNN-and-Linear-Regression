```python
import pandas as pd
import math
```

```python
transactions =  pd.read_excel("datasets/KPMG_VI_New_raw_data_update_final.xlsx",sheet_name = 1,header = 1)
transactions.drop(['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15',
       'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19',
       'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23',
       'Unnamed: 24', 'Unnamed: 25'],axis=1,inplace=True)
mean_columns = ['standard_cost','product_first_sold_date']
mode_columns = ['online_order','brand','product_line','product_class','product_size']
for i in mode_columns:
    if transactions[i].isnull().sum():
        transactions[i].fillna(transactions[i].mode()[0],inplace=True)
for i in mean_columns:
    if transactions[i].isnull().sum():
        transactions[i] = transactions[i].fillna(math.ceil(transactions[i].mean()))
transactions.isnull().sum()
```

```python
newcustomerlist = pd.read_excel("datasets/KPMG_VI_New_raw_data_update_final.xlsx",sheet_name=2,header=1)
newcustomerlist.drop(['Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
       'Unnamed: 19', 'Unnamed: 20'],axis=1,inplace=True)
cus_columns = ['last_name','DOB','job_title', 'job_industry_category']
for i in cus_columns:
    if newcustomerlist[i].isnull().sum():
        newcustomerlist[i].fillna(newcustomerlist[i].mode()[0],inplace=True)
newcustomerlist.isnull().sum()
```

```python
import math
customerdemographic = pd.read_excel("datasets/KPMG_VI_New_raw_data_update_final.xlsx",sheet_name=3,header=1)
customerdemographic.drop(['Unnamed: 13', 'Unnamed: 14',
       'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
       'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22',
       'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'],axis=1,inplace=True)
customerdemographic.drop('default',axis=1,inplace=True)
demo = ['last_name','DOB','job_title','job_industry_category','owns_car']
for columns in demo:
    if customerdemographic[columns].isnull().sum():
        customerdemographic[columns].fillna(customerdemographic[columns].mode()[0],inplace=True)
customerdemographic['tenure'].fillna(math.ceil(customerdemographic['tenure'].mean()),inplace=True)
customerdemographic.isnull().sum()
customerdemographic
```

```python
customeraddress = pd.read_excel("datasets/KPMG_VI_New_raw_data_update_final.xlsx",sheet_name=4,header=1)
customeraddress.drop(['Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8',
       'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12',
       'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16',
       'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20',
       'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24',
       'Unnamed: 25'],axis=1,inplace=True)
customeraddress.isnull().sum()
customeraddress
```

```python
# Joining customerdemographics and customeraddress

demographic = customerdemographic
demo = pd.merge(demographic,customeraddress,on='customer_id')
demo['gender'].replace({'F':'Female','M':'Male','Femal':'Female'},inplace=True)
demo
```

```python
import numpy as np
demo_cus = newcustomerlist.drop(['Rank','Value'],axis=1)
demo_cus.columns
add_column = {'customer_id':np.arange(4001,4001+len(demo_cus))}
demo_cus['customer_id'] = add_column['customer_id']
demonstrated = pd.concat([demo,demo_cus],axis=0)
demonstrated
```

```python
final_dataset = pd.merge(transactions,demo,on='customer_id')
final_dataset = final_dataset.sort_values(by='transaction_id')
final_dataset['first_name'] = final_dataset['first_name'].replace({'L;urette':'Lurette'})
final_dataset['state'] = final_dataset['state'].replace({'VIC':'Victoria','NSW':'New South Wales'})
for i in range(len(final_dataset['job_title'])):
    if ' I' in final_dataset.loc[i,'job_title']:
        final_dataset.loc[i,'job_title'] = final_dataset.loc[i,'job_title'].split(' I')[0]
    elif ' II' in final_dataset.loc[i,'job_title']:
        final_dataset.loc[i,'job_title'] = final_dataset.loc[i,'job_title'].split(' II')[0]
    elif ' III' in final_dataset.loc[i,'job_title']:
        final_dataset.loc[i,'job_title'] = final_dataset.loc[i,'job_title'].split(' III')[0]
    elif ' IV' in final_dataset.loc[i,'job_title']:
        final_dataset.loc[i,'job_title'] = final_dataset.loc[i,'job_title'].split(' IV')[0]
    else:
        pass
final_dataset
```

```python
final_dataset.to_excel('datasets/task1.xlsx')
```
